```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.interpolate import LinearNDInterpolator
import os

from astropy.table import Table
import astropy.units as u

from snewpy import model_path
from snewpy.neutrino import Flavor
from snewpy.models.ccsn import Analytic3Species

from asteria.simulation import Simulation
from asteria import set_rcparams

from snewpy.models.extended import ExtendedModel
from snewpy.models.base import SupernovaModel
from snewpy.neutrino import Flavor
from snewpy.models.ccsn import Mori_2023
import snewpy

from glob import glob
from scipy.interpolate import PchipInterpolator
import sys

import matplotlib
import snewpy.models.ccsn as ccsn
import scipy.stats as st
from astropy.table import Table, Column, vstack
from astropy.io import *
%matplotlib inline
set_rcparams()
from cycler import cycler
import emcee
import corner
from matplotlib import rcParams
matplotlib.rcParams['axes.prop_cycle'] = cycler("color", ['#ff355e', '#fd5b78', '#ff6037', '#ff9966', '#ff9933', '#ffcc33', '#ffff66', '#ccff00', '#66ff66', '#aaf0d1', '#50bfe6', '#16d0cb', '#6a1b9a', '#ff00cc' ])
from math import lgamma
from scipy.special import logsumexp
```

```python
distanceexp=1 * u.kpc
lowEexp=0*u.MeV
highEexp=100*u.MeV
Edeltexp=1*u.MeV
lowTexp=-0.29521*u.s
highTexp=0.29479*u.s
Tdeltexp=1*u.ms
mix_schemeexp='AdiabaticMSW'
hierarchyexp='normal'

expmod= {
    'name': 'Mori_2023',
    'param': {
        'axion_mass':100*u.MeV, 'axion_coupling':10*1e-10/u.GeV
    }
}
sim= Simulation(model=expmod,
                 distance=distanceexp, 
                 Emin=lowEexp, Emax=highEexp, dE=Edeltexp,
                 tmin=lowTexp, tmax=highTexp,  dt=Tdeltexp,
                 mixing_scheme=mix_schemeexp,
                 hierarchy=hierarchyexp)
sim.run()
```

```python
def log_likelihood_poisson(bg,scale,model_hits,data_hits):
    result=0
    for i in range(len(model_hits)):
        result+=data_hits[i]*np.log(bg+scale*model_hits[i])-bg-scale*model_hits[i]-lgamma(data_hits[i]+1)
    return result
```

```python
def compute_hits(dist,dt,sim):
    sim.rebin_result(dt)
    sim.scale_result(dist)
    t, hits = sim.detector_signal(dt)
        
    data=hits
    return data
```

```python
plt.figure()
plt.plot(compute_hits(1*u.kpc,10*u.ms,sim),color='b')
plt.plot(compute_hits(1*u.kpc,10*u.ms,sim)+1500)
plt.yscale('log')
plt.show()
```

```python
plt.figure()
plt.plot(compute_hits(1*u.kpc,10*u.ms,sim),color='b')
plt.plot(compute_hits(1*u.kpc,10*u.ms,sim)+1500)
plt.show()
```

```python
dt=1*u.ms
dist=1*u.kpc
data=compute_hits(1*u.kpc,10*u.ms,sim)+1500
model=compute_hits(1*u.kpc,10*u.ms,sim)
model_sim=sim

betamin=0
#betamax=30000/np.max(model)

betamax=4.5
    
self=model_sim.detector
bg_mean= self.i3_dom_bg_mu * dt.to(u.s).value * self.n_i3_doms + self.dc_dom_bg_mu * dt.to(u.s).value * self.n_dc_doms
bg_var= (self.i3_dom_bg_sig * np.sqrt(dt.to(u.s).value * self.n_i3_doms))**2 + (self.dc_dom_bg_sig * np.sqrt(dt.to(u.s).value * self.n_dc_doms))**2
bg_sig=np.sqrt(bg_var)
    
bmin= -10
#bmax= bg_mean+20*bg_sig

bmax=3000

scale_values=np.linspace(betamin,betamax,100)
bg_values=np.linspace(bmin,bmax,100)

scale_grid, bg_grid = np.meshgrid(scale_values, bg_values)
    
L=[[log_likelihood_poisson(b_, bet_, model, data) for bet_ in scale_values] for b_ in bg_values]


# Create the 2D plot
plt.figure(figsize=(7, 6))
contour = plt.contourf(scale_grid, bg_grid, L, cmap='viridis',levels=100)
cbar=plt.colorbar(contour)
cbar.set_label('Log Likelihood', fontsize=12)

# Add labels
plt.xlabel(r'Scale Factor ($\beta$)',fontsize=12)
plt.ylabel(r'Background (b)',fontsize=12)
plt.title(r'Example Log Likelihood',fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=10)

i, j = np.unravel_index(np.argmax(L), np.shape(L))
print(i)
print(j)
plt.axhline(bg_values[i], ls='--', color='k')
plt.axvline(scale_values[j], ls='--', color='k')

plt.show()
```
